## Cleaning & Preprocessing Dataset
---

```python
# Replacing null records with 0 and create a calculated field (GDP per capita)
df = df.fillna(0)
df['gdp_per_capita'] = np.where(df['population']!= 0, df['gdp']/df['population'], 0)
```

```python
#Make Dataframe Interactive
idf = df.interactive()
```

## Setting up Visualizations for the Dashboard
---

### 1<sup>st</sup> Viz Widgets

```python
# Creating Panel Slider
year_slider = pn.widgets.IntSlider(name='Year Slider', start=1750, end=2020, step=5,value=1850)
year_slider
```

```python
# Creating Panel Radio Buttons
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name='Y axis',
    options=['co2', 'co2_per_capita',],
    button_type='success'
)
```

#### 1<sup>st</sup> Viz: CO<sub>2</sub> Emissions by Continent

```python
continents = ['World', 'Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_pipelines = (
    idf[
        (idf.year <= year_slider) &
        (idf.country.isin(continents))
    ]
    .groupby(['country', 'year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
)
```

```python
co2_pipelines
```

```python
co2_plot = co2_pipelines.hvplot(x='year', by='country', y=yaxis_co2, line_width=2, title="CO2 emission by continent")
co2_plot
```

### 2<sup>nd</sup> Viz: Table of CO2 emissions over time by continent
---

```python
co2_table = co2_pipelines.pipe(pn.widgets.Tabulator, pagination='remote', page_size=10, sizing_mode='stretch_width')
co2_table
```

### 3<sup>rd</sup> Viz: C02 vs GDP Scatterplot
---

```python
co2_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.year == year_slider) &
        (~ (idf.country.isin(continents)))
    ]
    .groupby(['country', 'year', 'gdp_per_capita'])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
)
```

```python
co2_vs_gdp_scatterplot_pipeline
```

```python
co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(x='gdp_per_capita', y='co2', by='country', size=80, kind="scatter",title='CO2 Source Levels in each Country',
                                                                alpha=0.7, legend=False, height=500, width=500)
co2_vs_gdp_scatterplot
```

### 4<sup>th</sup> Viz Widget
---

```python
yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name='Y axis',
    options=['coal_co2', 'oil_co2', 'gas_co2'],
    button_type='success'
)

continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']
```

```python
co2_source_bar_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.country.isin(continents_excl_world))
    ]
    .groupby(['year', 'country'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
)
```

#### 4<sup>th</sup> Viz: CO<sub>2</sub> Source Levels in each continent

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(kind='bar', x='country', y=yaxis_co2_source, title='CO2 Source Levels in each Continent')
co2_source_bar_plot
```

## Building Dashboard from Vizzes
---

```python
template = pn.template.FastListTemplate(
    title='World CO2 Emissions Dashboard',
    sidebar=[
        pn.pane.Markdown("# CO2 Emissions and Climate Change"),
        pn.pane.Markdown("#### Carbon dioxide emissions are the primary driver of global climate change. It's widely recongised that to avoid global warming, we must find ways to reduce our carbon footprint."),
        pn.pane.JPG('orb.jpeg', sizing_mode='scale_both'),
        pn.pane.Markdown("## Settings"),
        year_slider
    ],
    main=[
        pn.Row(
            pn.Column(yaxis_co2, co2_plot.panel(width=700), margin=(0, 25)), 
            co2_table.panel(width=500)
        ),
        pn.Row(
            pn.Column(co2_vs_gdp_scatterplot.panel(width=600), margin=(0, 25)), 
            pn.Column(yaxis_co2_source, co2_source_bar_plot.panel(width=600))
        )
    ],
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)

# Display template
template.servable()
```
